## üìÜ 1. Vytvorenie d√°tumov a ƒçasov

```julia
# Vytvorenie d√°tumov a ƒçasov
using Dates
println(now())         # Aktu√°lny d√°tum a ƒças
println(today())       # Aktu√°lny d√°tum bez ƒçasu
```

```julia
# Vytvorenie konkr√©tneho d√°tumu
datum_vzor = Date(2025, 2, 19)  # 19. febru√°r 2025
println(datum_vzor)

# Vytvorenie konkr√©tneho ƒçasu
cas_vzor = Time(14, 30, 0)   # 14:30:00 (hodiny, min√∫ty, sekundy)
println(cas_vzor)

# Vytvorenie d√°tumu a ƒçasu
dt = DateTime(2025, 2, 19, 14, 30, 0)  # 18. febru√°r 2025, 14:30:00
println(dt)
```

## ‚è≥ 2. V√Ωpoƒçty s d√°tumami a ƒçasmi

```julia
# V√Ωpoƒçty s d√°tumami a ƒçasmi
# Pridanie alebo odƒç√≠tanie dn√≠, mesiacov, rokov
dt = DateTime(2025, 2, 19, 14, 30, 0) 

println(dt + Day(5))     # Prida≈• 5 dn√≠
println(dt - Month(2))  # Odƒç√≠ta≈• 2 mesiace
println(dt + Year(1))    # Prida≈• 1 rok

# Rozdiel medzi dvoma d√°tumami
d1 = Date(2025, 2, 19)
d2 = Date(2024, 12, 31)
println(d1 - d2)  # Rozdiel v d≈àoch (vr√°ti objekt typu Period)
println(typeof(d1 - d2))

```

```julia
# Rozdiel medzi dvoma d√°tumami v r√¥znych form√°toch
s1 = "01-01-2025"
s2 = "19-02-2025"

d1 = Date(s1, "dd-mm-yyyy")
d2 = Date(s2, "dd-mm-yyyy")

rozdiel = d2 - d1

println("Rozdiel je $(Day(rozdiel).value) dn√≠")  # Rozdiel je 49 dn√≠
```

## ‚õèÔ∏è 3. Extrahovanie ƒçast√≠ d√°tumu a ƒçasu

```julia
# Extrahovanie ƒçast√≠ d√°tumu a ƒçasu
dt = DateTime(2025, 2, 19, 14, 30, 0) 

println(year(dt))       # Rok (2025)
println(month(dt))      # Mesiac (2)
println(day(dt))        # De≈à (19)
println(hour(dt))       # Hodina (14)
println(minute(dt))     # Min√∫ta (30)
println(second(dt))     # Sekunda (0)
```

## üëâ 4. Form√°tovanie d√°tumu a ƒçasu

```julia
# Form√°tovanie d√°tumu a ƒçasu
dt = DateTime(2025, 2, 19, 14, 30, 0) 

# Konverzia na re≈•azec s vlastn√Ωm form√°tom
println(Dates.format(dt, "yyyy-mm-dd HH:MM:SS"))  # "2025-02-18 14:30:00"
println(Dates.format(dt, "d.mm.yyyy"))         # "18.02.2025"
println(typeof(dt))

println("------------------------------")
# Konverzia zo re≈•azca na d√°tum
dt2 = DateTime("2025-02-18 14:30:00", "yyyy-mm-dd HH:MM:SS")
println(dt2)
println(typeof(dt2))
```

## üìä 5. Urƒçenie d≈àa v t√Ω≈ædni a roku

```julia
# Urƒçenie d≈àa v t√Ω≈ædni
dt = DateTime(2025, 2, 19, 14, 30, 0) 

println(dayofweek(dt))      # 2 (utorok, kde pondelok = 1)
println(dayname(dt))        # "Tuesday"

println("------------------------------")

println(isleapyear(2024))  # true
println(isleapyear(2025))  # false

```

## üïí 6. Pr√°ca s ƒçasov√Ωmi p√°smami

```julia
## Pr√°ca s ƒçasov√Ωmi p√°smami
import Pkg 
Pkg.add("TimeZones")
using TimeZones

utc_now = ZonedDateTime(now(), tz"UTC")
vienna_now = ZonedDateTime(now(), tz"Europe/Vienna")  # V rovnakom p√°sme ako Bratislava

println(utc_now)
println(vienna_now)

println("------------------------------")

println(TimeZones.all_timezones())  # V√Ωpis v≈°etk√Ωch dostupn√Ωch ƒçasov√Ωch p√°siem

println("------------------------------")

bratislava_tz = TimeZone("Europe/Bratislava", TimeZones.Class(:LEGACY) | TimeZones.Class(:STANDARD))
bratislava_now = ZonedDateTime(now(), bratislava_tz)

println(bratislava_now)
```

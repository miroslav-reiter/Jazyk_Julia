```julia
using Logging, Dates

struct CustomLogger <: AbstractLogger
    stream::IO
    min_level::LogLevel
end

# Mus√≠me definova≈• min_enabled_level
Base.CoreLogging.min_enabled_level(logger::CustomLogger) = logger.min_level

# Funkcia urƒçuj√∫ca, ƒçi sa m√° logova≈•
function Base.CoreLogging.shouldlog(logger::CustomLogger, level, _module, group, id)
    return level >= logger.min_level
end

# Funkcia na spracovanie logovanej spr√°vy
function Base.CoreLogging.handle_message(logger::CustomLogger, level, message, _module, group, id, file, line)
    println(logger.stream, "[", level, "] ", now(), " - ", message)
end

# Nastavenie glob√°lneho loggera
global_logger(CustomLogger(stdout, Logging.Info))

# Test logovania
@info "Program sa spustil √∫spe≈°ne."
```

```julia
using Logging, Dates

struct CustomLogger <: AbstractLogger
    stream::IO
    min_level::LogLevel
end

# Definovanie povolenej √∫rovne logovania
Base.CoreLogging.min_enabled_level(logger::CustomLogger) = logger.min_level

# Overenie, ƒçi sa m√° spr√°va logova≈•
function Base.CoreLogging.shouldlog(logger::CustomLogger, level, _module, group, id)
    return level >= logger.min_level
end

# Spracovanie logovanej spr√°vy (opraven√©, akceptuje kwargs)
function Base.CoreLogging.handle_message(logger::CustomLogger, level, message, _module, group, id, file, line; kwargs...)
    println(logger.stream, "[", level, "] ", now(), " - ", message)
    for (k, v) in kwargs
        println(logger.stream, "  $k = $v")  # V√Ωpis dodatoƒçn√Ωch √∫dajov
    end
end

# Nastavenie vlastn√©ho loggera
global_logger(CustomLogger(stdout, Logging.Info))

# Testovanie logovania s parametrami
@info "Program sa spustil √∫spe≈°ne." ƒças = now()
```

## 4Ô∏è‚É£ Pou≈æitie modulu Printf
- Poskytuje form√°tovan√Ω v√Ωstup podobn√Ω jazyku C (printf).
- Pou≈æ√≠va sa na prec√≠zne form√°tovanie ƒç√≠sel a textu.

```julia
using Printf

@printf("Desatinn√© ƒç√≠slo: %.2f\n", 3.14159)  # V√Ωstup: Desatinn√© ƒç√≠slo: 3.14
@printf("ƒå√≠slo: %d, Text: %s\n", 42, "Julia")  # V√Ωstup: ƒå√≠slo: 42, Text: Julia
```

### Pou≈æitie @sprintf na ulo≈æenie form√°tovan√©ho textu do premennej
- Na rozdiel od @printf, @sprintf vracia re≈•azec namiesto priameho v√Ωpisu.

```julia
# Pou≈æitie @sprintf na ulo≈æenie form√°tovan√©ho textu do premennej
using Printf

@printf("%s: %.3f\n", "Pi", 3.14159)  # V√Ωstup: Pi: 3.142
formatted_string = @sprintf("%s: %.3f", "Pi", 3.14159)
println(formatted_string)  # V√Ωstup: Pi: 3.142
```

```julia
pi_value = 3.14159
formatted_string = "Pi: $pi_value"
println(formatted_string)  # V√Ωstup: Pi: 3.14159
```

```julia
formatted_string = "Pi: $(round(3.14159, digits=3))"
println(formatted_string)  # V√Ωstup: Pi: 3.142
```

## üìÇ 1. Otvorenie a ƒç√≠tanie s√∫boru
- Pou≈æitie funkcie read, readline alebo readlines na naƒç√≠tanie obsahu

```julia
# Otvorenie a ƒç√≠tanie s√∫boru

# Naƒç√≠tanie cel√©ho s√∫boru ako re≈•azec
obsah = read("subor.txt", String)
println(obsah)

# Naƒç√≠tanie prvej riadky
s_prva_riadka = readline("subor.txt")
println(s_prva_riadka)

# Naƒç√≠tanie v≈°etk√Ωch riadkov do poƒæa
riadky = readlines("subor.txt")
println(riadky)
```

```julia
try
    obsah = read("subor.txt", String)
    println("Obsah s√∫boru: ", obsah)
catch e
    println("Chyba: ", e)
end
```

```julia
# Overenie, ƒçi s√∫bor existuje
if isfile("subor.txt")
    obsah = read("subor.txt", String)
    println(obsah)
else
    println("S√∫bor 'subor.txt' neexistuje. Vytv√°ram nov√Ω...")
    write("subor.txt", "Toto je testovac√≠ obsah s√∫boru.\n")
end

# Teraz s√∫bor existuje, m√¥≈æeme ho naƒç√≠ta≈•
obsah = read("subor.txt", String)
println("Obsah s√∫boru: ", obsah)
```

```julia
while true
    try
        # Pokus o naƒç√≠tanie s√∫boru
        obsah = read("subor.txt", String)
        println("Obsah s√∫boru: ", obsah)
        break  # Ak sa podar√≠ preƒç√≠ta≈•, preru≈°√≠ cyklus
    catch e
        # Ak s√∫bor neexistuje, vytvor√≠ sa nov√Ω a cyklus pokraƒçuje
        println("S√∫bor 'subor.txt' neexistuje alebo je probl√©m s ƒç√≠tan√≠m. Vytv√°ram nov√Ω...")
        try
            write("subor.txt", "Toto je testovac√≠ obsah s√∫boru.\n")
            println("S√∫bor bol √∫spe≈°ne vytvoren√Ω.")
        catch ew
            println("Chyba pri z√°pise do s√∫boru: ", ew)
            break  # Ak zlyh√° aj z√°pis, preru≈°√≠ sa cyklus, aby sa nebe≈æal donekoneƒçna
        end
    end
end
```

## üìù 2. Z√°pis do s√∫boru
- Pou≈æitie funkcie write alebo open s re≈æimom "w"

```julia
# Jednoduch√Ω z√°pis do s√∫boru
write("vystup.txt", "Toto je testovac√≠ z√°pis do s√∫boru.\n")

# Z√°pis viacer√Ωch riadkov
open("vystup.txt", "w") do f
    write(f, "Prv√Ω riadok\n")
    write(f, "Druh√Ω riadok\n")
end

# ƒå√≠slo 14 znamen√° poƒçet zap√≠san√Ωch bajtov pri prvom pou≈æit√≠ write()
# Re≈•azec "Toto je testovac√≠ z√°pis do s√∫boru.\n" obsahuje 38 znakov (vr√°tane \n).
# Ak je s√∫bor k√≥dovan√Ω v UTF-8, poƒçet bajtov m√¥≈æe by≈• r√¥zny v z√°vislosti od diakritiky.
# Pri ASCII znakoch plat√≠, ≈æe ka≈æd√Ω znak zaber√° 1 bajt.
# Ak namiesto "Toto je testovac√≠ z√°pis do s√∫boru.\n" pou≈æije≈° kr√°tky re≈•azec, napr√≠klad "Hello\n", vyp√≠≈°e men≈°ie ƒç√≠slo (napr. 6).
```

## üìå 3. Pridanie textu do existuj√∫ceho s√∫boru
- Pou≈æitie re≈æimu "a" (append)

```julia
# Pridanie textu do existuj√∫ceho s√∫boru

open("vystup.txt", "a") do f
    write(f, "Tret√≠ riadok (doplnen√Ω dodatoƒçne).\n")
end
```

## üîç 4. Kontrola existencie s√∫boru

```julia
# Kontrola existencie s√∫boru
if isfile("subor.txt")
    println("S√∫bor existuje.")
else
    println("S√∫bor neexistuje.")
end
```

```julia
# Odstr√°nenie s√∫boru

rm("vystup.txt")
```

## üìÇ 6. Pr√°ca s adres√°rmi
- readdir() ‚Äì z√≠skanie zoznamu s√∫borov v prieƒçinku
- mkdir() ‚Äì vytvorenie adres√°ra
- rm(dir, recursive=true) ‚Äì odstr√°nenie prieƒçinka

```julia
# Pr√°ca s adres√°rmi

# V√Ωpis obsahu aktu√°lneho prieƒçinka
println(readdir("."))

# Vytvorenie adres√°ra
mkdir("novy_priecinok")

# Odstr√°nenie adres√°ra
rm("novy_priecinok", recursive=true)
```
